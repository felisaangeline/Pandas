# Plotting dari suatu Pandas Data Frame

### Import Modul dan Cek Versi

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

### Persiapan Data Frame

```python
n_rows = 40
n_cols = 5 
cols = tuple ('ABCDE')
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

# Melakukan line plot

```python
df.head()
```

```python
df.plot(kind='line')

#Setiap garis akan merepresentasikan setiap kolom
#Sumbu x merepresentasikan baris
#sumbu y merepresentasikan nilai yang tertampung dalam Data Frame
```

```python
df[['A', 'B']].plot(kind='line')
```

# Bar plot

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A', 'B']].plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='barh')

#merepresentasikan bar secara horizontal
```

# Area plot
