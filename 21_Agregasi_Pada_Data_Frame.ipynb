# Agregasi pada data frame menggunakan agg()

Agregasi pada data frame menggunakan agg()

```python
import pandas as pd
print(pd.__version__)
```

Persiapan Data Frame

```python
df = pd.read_csv('/content/iris.csv')
df
```

```python
df.groupby('Species')['PetalLengthCm'].count().to_frame()

#count untuk menghitung jumlah baris
#to_frame() hasil output ditampilkan dalam bentuk frame
#fungsi agregrasinya dalam bentuk count()
```

```python
df.groupby('Species')['PetalLengthCm'].mean().to_frame()

#fungsi agregasinya adalah mean() yang berfungsi untuk menghitung nilai rata-rata
```

```python
df.groupby('Species')['PetalLengthCm'].median().to_frame()
```

# Agregasi dengan agg()

```python
df.groupby('Species')['PetalLengthCm'].agg(['count', 'mean', 'median'])

#fungsi agg() dalam bentuk list yang berisi fungsi agregasi apa saja yang mau dikenakan pada data yang sudah dikelompokkan
```

# Agregasi dengan describe()

```python
df.groupby('Species')['PetalLengthCm'].describe()
```
