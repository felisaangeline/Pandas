# Menata ulang Data Frame dengan multiple indexes menggunakan unstack

Import Modul dan Cek Versi

```python
import pandas as pd

print(pd.__version__)
```

Persiapan Data Frame

```python
df = pd.read_csv("/content/titanicfull.csv")
df.head()
```

# Data Frame dengan multiple indexes dari hasil groupping

```python
df.groupby(['sex', 'pclass'])['survived'].mean().to_frame()

#data survived akan diamati
#frame yang dihasilkan terdapat 2 buah index yaitu index sex dan pclass
#terdapat 1 kolom yaitu survived
```

# Menata ulang data frame dengan multiple indexes

```python
df.groupby(['sex', 'pclass'])['survived'].mean().unstack()

#data survived akan diamati
#aggregate yang digunakan yaitu mean
#terdapat 2 index yaitu sex dan pclass
#penempatan index yang berbeda
#perbedaannya adalah dari cara representasinya
```
