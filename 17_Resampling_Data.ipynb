# Resampling pada data deret waktu (time series data)

### Import Modul dan Cek Versi

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 365 * 24 

# 365 merepresentasikan jumlah hari dalam satu tahun
# 24 merepresentasikan jumlah jam dalam satu hari
# akan membentuk deret waktu

n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols) #data bernilai integer dengan jangkauan 1-20

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
#index data frame akan diisi dengan deret waktu, jumlah index waktu yang di generate yaitu n_rows

df
```

# Resampling data dengan interval monthly

```python
df.resample('M')['col1'].sum().to_frame()

# parameter M berarti melakukan resampling berbasis bulan
# data yang akan diamati yaitu col1
# .to_frame() mengubah tampilan menjadi frame agar lebih cantik
```

# Resampling data dengan interval daily

```python
df.resample('D')['col1'].sum().to_frame()

# parameter D berarti melakukan resampling berbasis hari
# data yang akan diamati yaitu col1
```
