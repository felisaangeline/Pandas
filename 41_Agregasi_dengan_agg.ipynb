# Agregasi pada banyak kolom Data Frame dengan agg

### Import Modul dan Cek Versi

```python
import pandas as pd

print(pd.__version__)
```

### Persiapan data frame

```python
df = pd.read_csv('/content/titanicfull.csv')
df.head()
```

```python
df.groupby('pclass').agg({'pclass': 'count',
                        'age': ['mean', 'max'],
                        'survived': 'mean'})

#fungsi agg akan menerima suatu dictionary
#data dictionary berisi keys dan values
#jika ingin mengenakan lebih dari 1 fungsi agregasi maka harus mengenakan [ ]
```

```python
df.groupby('pclass').agg(n_pass=('pclass', 'count'), 
                         avg_age=('age', 'mean'), 
                         max_age=('age', 'max'),
                         survival_rate=('survived', 'mean'))

#setiap key word akan merepresentasikan nama kolom
```
